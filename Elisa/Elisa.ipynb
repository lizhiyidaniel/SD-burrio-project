```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
listings = pd.read_csv("datasets/listings.csv")
burritos = pd.read_csv("datasets/burritos.csv")
```

```python
listing_cols = ["id", "neighbourhood", "price"]
listings_df = listings[listing_cols].copy()
listings_df = listings_df.rename(columns = {"id": "ID", "neighbourhood":"Neighborhood", "price": "Price"})

listings_df.head()
```

```python
listings_grouped = listings_df.groupby("Neighborhood")
average_price = listings_grouped["Price"].median()
listings_count = listings_grouped.count()["ID"]

listing_summary_df = pd.DataFrame({"Total Listings": listings_count, "Average Price/Night": average_price}).sort_values("Average Price/Night", ascending = False)

listing_summary_df
```

```python
burrito_cols = ["Location", "Burrito", "Neighborhood", "Yelp", "Google", "Cost", "overall"]
burrito_df = burritos[burrito_cols].dropna().copy()
burrito_df.head()
```

```python
summary_df = pd.merge(burrito_df, listing_summary_df, how="inner", on=["Neighborhood"])

summary_df.dropna().sort_values("overall", ascending = False)

```

```python
summary_df.to_csv("summary_data.csv")
```
