```python
import pandas as pd
```

```python
file = '../datasets/listings.csv'
```

```python
data = pd.read_csv(file)
```

```python
data.head()
```

```python
data = data[["id","neighbourhood"]]
```

```python
data.head()
```

```python
data = data.rename(columns ={'neighbourhood':'neighborhood'})
```

```python
listing_group = data.groupby('neighborhood')
listing_count = listing_group.count()['id']

listing = pd.DataFrame({"listings":listing_count})
```

```python
listing
```

```python
listing = listing.sort_values(by='listings',ascending = True)
```

```python
listing
```

```python
listing.to_csv('airbnb_listing.csv')
```
